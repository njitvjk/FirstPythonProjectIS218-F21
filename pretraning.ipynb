```python

from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import DataCollatorForLanguageModeling, TrainingArguments, Trainer
from datasets import Dataset

# 1. dataset for demo
texts = [
       "Climate change is a pressing global issue.",
        "Renewable energy sources are essential for sustainability.",
        "Advances in quantum computing are accelerating technology.",
        "Space exploration expands our understanding of the universe.",
        "Healthcare innovation improves quality of life worldwide."
]
dataset = Dataset.from_dict({"text": texts})

# 2. Load model & tokenizer
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(model_name)

# 3. Tokenization
def tokenize(example):
    return tokenizer(example["text"], padding="max_length", truncation=True, max_length=64)

tokenized_dataset = dataset.map(tokenize, batched=True)

# 4. Data collator for causal language modeling
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

# 5. Training arguments
training_args = TrainingArguments(
    output_dir="./gpt2-finetuned",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    logging_steps=10,
    save_steps=10,
    save_total_limit=2,
    report_to="none"
)

# 6. Trainer setup
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator
)

# 7. Train the model
trainer.train()

# 8. Inference (generate text)
prompt = "AI is"
input_ids = tokenizer(prompt, return_tensors="pt").input_ids
output = model.generate(input_ids, max_length=20)
print("=== Generated Text ===")
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
